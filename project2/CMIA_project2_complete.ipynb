# Building a Trading Signal Using the Indicator

Using the indicator that we have created, we then try to create a trading signal.

We start by thinking how we can trade this strategy.

The most simplistic use of the indicator to trade would be using the indicator as the trading signal without any transformation. In our example, that would mean buy (go long) when the change in GDP is positive, and sell (go short) when the change in GDP is negative. This approach is shown in basic_signal() below.

We encourage you to try some new signals. As an example of how a signal might look like, you can look at the momentum trading signal in the Software Engineering project (Project 1).

```python
def basic_signal(ts):
    # ts is a vector
    signal = pd.DataFrame(np.zeros((len(ts.index), 1)),index=ts.index, columns=['signal'])
    signal[ts > 0] = 1
    signal[ts < 0] = -1
    
    return signal
```

```python
ca_sgnl = basic_signal(ca_gdp)
us_sgnl = basic_signal(us_gdp)
```

### Calculate PnL

```python
def calculate_pnl(ret, pos):
    # function to calculate profit and loss
    # we trade at the end of the period, and take profit & loss on the market move next period on our position
    # that's why we are shifting positon vector by one
    return ret.multiply(pos.shift(), axis=0).dropna()
```

```python
# build the return series for each of the series
ca_returns = asset_prices[ca_assets].pct_change()
us_returns = asset_prices[us_assets].pct_change()

# Resample the Signal into daily signal
ca_daily_pos = ca_sgnl.resample('D').ffill()
us_daily_pos = us_sgnl.resample('D').ffill()
```

```python
us_pnl = calculate_pnl(us_returns, us_daily_pos['signal'])
ca_pnl = calculate_pnl(ca_returns, ca_daily_pos['signal'])

# bug fix: replace us_daily_pos by ca_daily_pos
# ca_pnl = calculate_pnl(ca_returns, us_daily_pos['signal'])
```

#### Thoughts
- Equity and Currency are two different asset classes. \
 They have different exposures and styles. 
 Therefore, it's inappropriate to use the same set of signal for both types of assets).
 
 Additionaly, if we plot the cumulative return of the assets (showing above),\
 we notice that in general currency indices are expected to have zero return\
 while stock indices are expected to have positive return over long run and with strong momentum.
 
 This leads to the idea of avoiding short position for equity indices and try time series momentum indicator


- Macro data is often lagged, we might use 1. change in actual(ex), 2. surprise value(ex), as (actual - expected)


- For cross-sectional signal, we could try the idea of pair-trading (no time for this)

#### Add surprise_value and change_value
- surprise_value = actual - expected
- change_value = actual_value - actual_value

```python
for df_name, _df in data.items():
    if df_name != 'asset_prices':
        _df['surprise'] = _df['actual_value'] - _df['expected_value']
        _df['change_actual'] = _df['actual_value'] - _df['actual_value'].shift()
```

#### Define suprise signal
1. macro data is different from expected\
if actual >> expected / surprise is positive, market moves upward, vise versa


2. people expect a trend of economy going up/down (change in actual)\
By default, we assume a long position since in the long run equity is expected to yield positive return and no matter under negative change_actual or negative_surprise, average return of EQ assets are still POSITIVE. We hold no position when there is negative surprise


```python
def surprise_signal(input_df):
    """
    input_df is the input macro dataframe with columns: [actual_value  expected_value  surprise, change_actual]
    when surprise < 0 => no signal
    all other situations: long signal
    return: trading signal df of the same index as the input_df
    """
    signal = pd.DataFrame(np.ones((len(input_df.index), 1)),index=input_df.index, columns=['signal'])

    signal[(input_df['surprise'] < 0)] = 0

    return signal
```

```python
# check performance
# only apply the strategy on Equity asset class - no time for FX
ca_eq = ['PT1 Index']
us_eq = ['ES1 Index']

# get daily data
ca_eq_returns = asset_prices[ca_eq].pct_change()
us_eq_returns = asset_prices[us_eq].pct_change()

us_gdp_signal = surprise_signal(data['US_GDP'])
ca_gdp_signal = surprise_signal(data['CA_GDP'])

ca_gdp_daily_pos = ca_gdp_signal.resample('D').ffill()
us_gdp_daily_pos = us_gdp_signal.resample('D').ffill()

us_gdp_pnl = calculate_pnl(us_eq_returns, us_gdp_daily_pos['signal'])
ca_gdp_pnl = calculate_pnl(ca_eq_returns, ca_gdp_daily_pos['signal'])

cum_gpd_return_us = (us_gdp_pnl + 1.0).cumprod() - 1.0
us_gdp_trading_sharpe = sharpe_calc(us_gdp_pnl) * np.sqrt(252)

cum_gpd_return_ca = (ca_gdp_pnl + 1.0).cumprod() - 1.0
ca_gdp_trading_sharpe = sharpe_calc(ca_gdp_pnl) * np.sqrt(252)

df_sharpe = pd.DataFrame({
    "CA Baseline": ca_baseline_sharpe,
    "CA Trading": ca_trading_sharpe,
    "US Baseline": us_baseline_sharpe,
    "US Trading": us_trading_sharpe,
    "CA New": ca_gdp_trading_sharpe,
    "US New": us_gdp_trading_sharpe,
}).fillna("")
```
